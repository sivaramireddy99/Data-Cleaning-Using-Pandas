```python
df["Last_Name"]
```

```python
# STRIP :- REMOVE WHITESPACES  AND 
#l MEANS :- LEFT
#r MEANS :- RIGHT
df["Last_Name"].str.lstrip("...")
# LONG PROCESS METHOD AND TIME TAKEN PROCESS
```

```python
## SHORTEST METHOD FOR STRIP IS STRIP("123._/")
df["Last_Name"].str.lstrip("123._/")
```

```python
df.loc[:,"Last_Name"] = df["Last_Name"].str.strip("123._/")
## ACTUAL DATA FRAME
# LOC :- INDEXING TO LABELS
# HERE (:) :- ALL ROWS
```

```python
df
```

```python
df["Phone_Number"]
```

```python
df.loc[:,"Phone_Number"] = df["Phone_Number"].str.replace(r"[^0-9]","",regex = True)
df.loc[:,"Phone_Number"] = df["Phone_Number"].replace({"N/a": pd.NA, "":pd.NA})
## REPLACING THE ESCAPE CHARACTER WITH SPACE( )
```

```python
df
```

```python
df.loc[:,"Phone_Number"] = df["Phone_Number"].mask(df["Phone_Number"].str.len()!= 10, other=pd.NA)
##NOW NAN BECOME NA
```

```python
df
```

```python
df = df.dropna(subset=['Phone_Number'])
## NOW SIMPLING REMOVING THE UNNECESSARY ROWS
```

```python
df
```

```python
df["Address"]
```

```python
df["Address"].str.split(", ", n=2, expand=True)
### NOW ADDRESS SPLIT INTO COLUMNS
##SPLIT :- ONE COLUMN BECOMES TWO OR MORE COUMNS BASED ON CONDITION.
```

```python
df[["Street Address","State","Zip_Code"]] = df["Address"].str.split(",", n=2, expand=True)
### GIVING THE NAMES OF SPLITED COLUMNS. IF GET WARNING JUST LEAVE IT.
```

```python
df
```

```python
df = df.replace("N/a","")
```

```python
df = df.replace("NaN","")
```

```python
df.fillna('')
## NONE CELLS BECOMES EMPTY( )
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No","N")
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No","N")
```
